# WOIK Tour
Load sample payload and run the engine.


```python
from pathlib import Path
import json
import subprocess, sys
REPO = Path.cwd()
subprocess.check_call([sys.executable,'-m','pip','install','-e','.[vis]'], cwd=str(REPO))
payload = json.loads((REPO/'examples/data/sample_payload.json').read_text())
payload
```

```python
from woik.core.engine import WOIKEngine
from woik.plugins.builtin.leak_detector import LeakDetectorPlugin
from woik.plugins.builtin.pump_efficiency import PumpEfficiencyPlugin
from woik.plugins.builtin.quality_risk import QualityRiskPlugin
engine = WOIKEngine()
engine.register(LeakDetectorPlugin())
engine.register(PumpEfficiencyPlugin())
engine.register(QualityRiskPlugin())
report = engine.run(payload)
report.summary()
```
